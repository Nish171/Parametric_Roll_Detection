```python
# Dataset generator
n = 10
m = 5
hop = 0.25
input_dim   = 30 * n
pred_dim    = 30 * m
shift       = pred_dim
batch_size  = 16
Data        = DS(
                input_dim, 
                pred_dim, 
                shift, 
                skip=0.25, 
                hop=hop, 
#                 normalizer=1, 
                batch_size=batch_size,
                in_cols=['roll', 'pitch','heave','wave']
                )
```

```python
data = Data.get_sim_data(30, split=False)
```

```python
t = data.loc[:, 'time']
roll = data.loc[:, 'roll']
pitch = data.loc[:, 'pitch']
heave = data.loc[:, 'heave']
fs = 4
```

```python
win_size = 1024
over = 512
window = signal.windows.boxcar(win_size)
```

```python
pf, pt, pSxx = signal.spectrogram(pitch, fs, window=window, noverlap=over, scaling='spectrum')
rf, rt, rSxx = signal.spectrogram(roll, fs, window=window, noverlap=over, scaling='spectrum')
hf, ht, hSxx = signal.spectrogram(heave, fs, window=window, noverlap=over, scaling='spectrum')
```

```python
cut = 30

plt.figure(figsize=(15,30))
plt.subplot(611)
plt.pcolormesh(rt, rf[:cut], rSxx[:cut], shading='gouraud')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')

plt.subplot(612)
plt.plot(t, roll)
plt.xlabel("Time [sec]")
plt.ylabel('Roll amp [deg]')

# plt.figure(figsize=(15,20))
plt.subplot(613)
plt.pcolormesh(pt, pf[:cut], pSxx[:cut], shading='gouraud')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')

plt.subplot(614)
plt.plot(t, pitch)
plt.xlabel("Time [sec]")
plt.ylabel('Pitch amp [deg]')


plt.subplot(615)
plt.pcolormesh(ht, hf[:cut], hSxx[:cut], shading='gouraud')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')

plt.subplot(616)
plt.plot(t, heave)
plt.xlabel("Time [sec]")
plt.ylabel('Heave amp [mm]')
plt.show()
```

```python
fr, t, rSxx = signal.spectrogram(roll, fs)
plt.pcolormesh(t, fr, rSxx, shading='gouraud')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
