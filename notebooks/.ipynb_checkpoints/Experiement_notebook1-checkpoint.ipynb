```python
from src.data.Dataset import Dataset as DS
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
train_x_path = r'data\processed\Simulations_01\rp_30_n_4_m_2\train\train_X.csv'
train_y_path = r'data\processed\Simulations_01\rp_30_n_4_m_2\train\train_Y.csv'
```

```python
train_x = np.loadtxt(train_x_path, delimiter=',')
train_y = np.loadtxt(train_y_path, delimiter=',')
```

```python
train_x.shape
```

```python
train_y.shape
```

```python
train_x_ten = tf.convert_to_tensor(train_x)
train_y_ten = tf.convert_to_tensor(train_y)
```

```python
train_ds = tf.data.Dataset.from_tensor_slices((train_x_ten, train_y_ten))
```

```python
inp, target = next(iter(train_ds))
```

```python
t1 = range(0, len(inp))
t2 = range(len(inp), len(inp)+len(target))
```

```python
plt.figure(figsize=(15,10))
plt.plot(t1, inp)
plt.plot(t2, target)
plt.show()
```
